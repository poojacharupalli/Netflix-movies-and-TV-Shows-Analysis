# Netflix Movies and TV shows Analysis Project

### This project involves a comprehensive analysis of Netflix movies and TV shows data, exploring various aspects such as content types, release years, directors, countries, and genres. The analysis aims to provide insights into the Netflix catalog, highlighting trends, patterns, and notable features.

```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
```

## Data Loading and Preprocessing

```python
df = pd.read_csv("netflix_titles.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df['show_id'].value_counts().sum()
```

```python
df.set_index('show_id',inplace=True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df[['director', 'cast', 'country']] = df[['director', 'cast', 'country']].replace(np.nan, "Unknown ")
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(subset=['rating','duration','date_added'],inplace =True) 
```

```python
df.isna().sum()
```

## Data Exploration

```python
df['date_added']=df['date_added'].astype('datetime64[ms]')
```

```python
df.info()
```

```python
df.head()
```

```python
df['loading_Year'] = df ['date_added'].dt.year
df['loading_Year'] 
```

```python
df['loading_Month']= df ['date_added'].dt.month_name()
df['loading_Month']
```
